# Method 2 - Rows first then Columns (filling nan values based on mean of different classes)

```python
df = pd.read_csv("train.csv")
df.shape
```

```python
df_null_clm = df.isnull().sum().to_frame("null count")
df_null_clm["null percentage"] = df_null_clm["null count"] / df.shape[0] * 100
df_null_clm
```

```python
df_null_per = df_null_clm[df_null_clm["null percentage"] > 20]
df_null_per
```

```python
l = df_null_per.index
l
```

```python
df_drop_clm = df.drop(columns = l)
df_drop_clm.shape
```

```python
df_num = df_drop_clm.select_dtypes(include = ["int64", "float64"])
df_num.shape
```

```python
df1 = df_num.isnull().sum() / df_num.shape[0] * 100
df1
```

```python
l1 = df1[df1 > 0].keys()
l1
```

```python
l2 = df_drop_clm["LotConfig"].value_counts().keys()
l2
```

```python
df_drop_clm.loc[df_drop_clm["LotConfig"] == "Inside", "LotFrontage" : "LotConfig"]
```

```python
df_lot1 = df_drop_clm.loc[df_drop_clm["LotConfig"] == "Inside", "LotFrontage"]
df_lot1
```

```python
df_lot1 = df_lot1.replace(np.nan, df_lot1.mean())
df_lot1
```

```python
for i in l2:
    df_drop_clm.loc[df_drop_clm["LotConfig"] == i, "LotFrontage"] = df_drop_clm.loc[df_drop_clm["LotConfig"] == i, "LotFrontage"].replace(np.nan, df_drop_clm.loc[df_drop_clm["LotConfig"] == i, "LotFrontage"].mean())
```

```python
df_drop_clm["LotFrontage"]
```

```python
l3 = df_drop_clm["GarageQual"].value_counts().keys()
l3
```

```python
df_gar1 = df_drop_clm.loc[df_drop_clm["GarageQual"] == "TA", "GarageYrBlt"]
df_gar1
```

```python
for i in l3:
    df_drop_clm.loc[df_drop_clm["GarageQual"] == i, "GarageYrBlt"] = df_drop_clm.loc[df_drop_clm["GarageQual"] == i, "GarageYrBlt"].rep(np.nan, df_drop_clm.loc[df_drop_clm["GarageQual"] == i, "GarageYrBlt"].mean())
```

```python
for i in l3:
    df_drop_clm.loc[df_drop_clm["GarageQual"] == i, "GarageYrBlt"] = \
        df_drop_clm.loc[df_drop_clm["GarageQual"] == i, "GarageYrBlt"] \
        .fillna(df_drop_clm.loc[df_drop_clm["GarageQual"] == i, "GarageYrBlt"].mean())
```
