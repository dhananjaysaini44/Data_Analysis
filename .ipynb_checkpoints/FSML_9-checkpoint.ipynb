```python
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import OneHotEncoder
pd.set_option("display.max_rows", None)
pd.set_option("display.max_columns", None)
```

# Encoding

Encoding:- 
1. Nominal Encoding - Onehot Encoding (No Priority based)
2. Ordinal Encoding - Label Encoding (Priority Based)

```python
df = pd.read_csv("tips.csv")
df
```

```python
pd.get_dummies(df, dtype = int, drop_first = True)
```

```python
df1 = OneHotEncoder(sparse_output = False, drop = "first")
```

```python
df2 = df1.fit_transform(df[["sex", "smoker", "day", "time"]])
```

```python
pd.DataFrame(df2, columns = ["sex_Male", "smoker_Yes", "day_Sat", "day_Sun", "day_Thur", "time_Lunch"])
```

```python
df = sns.load_dataset("tips")
df
```

```python
df1 = df[["sex", "smoker", "day", "time"]]
df1
```

```python
df1["day"].unique()
```

```python
order = {'Sun':0, 'Sat':3, 'Thur':1, 'Fri':2}
```

```python
df1["day"].map(order)
```

```python
df1["day"] = df1["day"].map(order)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
ar = le.fit_transform(df1["time"])
```

```python
pd.DataFrame(ar, columns = ["time"])
```

```python
df1["time"] = pd.DataFrame(ar, columns = ["time"])
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories = [["Male", "Female"]])
```

```python
ar1 = oe.fit_transform(df1[["sex"]])
```

```python
pd.DataFrame(ar1, columns = ["sex"])
```

```python
df1["sex"] = pd.DataFrame(ar1, columns = ["sex"])
```

```python
df1
```

```python
oh = OneHotEncoder(sparse_output=False)
```

```python
ar3 = oh.fit_transform(df1[["smoker"]])
ar3
```

```python
df2 = pd.DataFrame(ar3, columns = ["no_smoker", "yes_smoker"])
```

```python
df1 = pd.concat([df1, df2], axis = 1)
df1
```

```python
df1.drop(columns = "smoker")
```
