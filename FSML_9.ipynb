1. What SHOULD be scaled:
   Integers & Floats: Standard numeric data (Age, Salary, Fare, Distance).
   Ordinal Encoded Features: These are features with a built-in rank (e.g., Education: $1$ for High School, $2$ for Masters).Because the model interprets $2$ as "larger" than $1$, you scale them so they are on the same magnitude as your other numbers.
2. What SHOULD NOT be scaled:
   Nominal Encoded (One-Hot) Features: These are your $0$s and $1$s created from categories with no order (e.g., City_NewYork, City_Tokyo). They are already "pre-scaled" to a $0-1$ range. Scaling them (especially with StandardScaler) can actually ruin their sparse meaning.
   Target Variables: Usually, you don't scale the column you are trying to predict ($y$), only the input features ($X$).

```python
from sklearn.compose import ColumnTransformer
```

```python
l6_scalableFeatures = ["pclass", "age", "sibsp", "parch", "fare", "embark_town"]
```

```python
ct = ColumnTransformer([
        ('my_scaler', StandardScaler(), l6_scalableFeatures)
    ], 
    remainder='passthrough'
)
```
